```python
import csv
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
from matplotlib import rc, font_manager

filename='csv/temperature_2014.csv'
font_name = font_manager.FontProperties(fname='C:\Windows\Fonts/GULIM.TTC').get_name()
rc('font', family=font_name)
df = pd.read_csv('csv/temperature_2014.csv')


with open(filename) as f:
    reader = csv.reader(f)
    head_row=next(reader)
    dates, highs,lows = [],[],[]
    for row in reader:
        try:
            current_date = datetime.strptime(row[0],"%Y-%m-%d")
            high = (int(row[1])-32)*5/9
            low = (int(row[3])-32)*5/9
        except ValueError:
            print(current_date,'missing date')
        else:
            dates.append(current_date)
            highs.append(high)
            lows.append(low)

# 
fig = plt.figure(dpi=128,figsize=(10,6))
#x date, y highs, red
plt.plot(dates, highs, c='red', alpha=0.5)
plt.plot(dates, lows, c='blue', alpha=0.5)
#   (highs~lows) , alpha=0.1  0.1
plt.fill_between(dates,highs,lows,facecolor='blue',alpha=0.1)
#  
plt.title(' ',fontsize=20)
plt.xlabel("",fontsize=16)
plt.ylabel("",fontsize=16)
plt.tick_params(axis='both',labelsize=16) #  
plt.savefig('chart.png',dpi=400,bbox_inches='tight')
plt.show()

print(":",round(df['Max TemperatureF'].corr(df['Min TemperatureF']),4))
```
