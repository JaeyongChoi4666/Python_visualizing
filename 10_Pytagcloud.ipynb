```python
from collections import Counter
from konlpy.tag import Okt
import pytagcloud
```

```python
file=open("news.txt", encoding='utf-8')
data=file.read()
print(data)
```

```python
npl = Okt()
nouns=npl.nouns(data)
print(nouns)
```

```python
count=Counter(nouns)
print(count)
```

```python
tag2=count.most_common(20)
taglist=pytagcloud.make_tags(tag2,maxsize=80)
print(taglist)
file.close()
```

```python
pytagcloud.create_tag_image(taglist,'wordcloud.jpg', size=(900,600), fontname='HYKANM', rectangular=False)
```
