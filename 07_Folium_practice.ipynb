```python
import csv
import folium
import numpy as np
import pandas as pd

file = open('csv/CCTV_20190917.csv','r',encoding="utf-8")
rdr = csv.reader(file)
next(rdr,None)
for idx,tmp in enumerate(rdr):
    if idx == 0: #최초 1번 객체 만들기, 시작위치의 중심이 location에 정의된 곳
        map_data = folium.Map(location=[float(tmp[10]),float(tmp[11])], zoom_size=13, zoom_start=13)
    folium.Marker([float(tmp[10]),float(tmp[11])],icon=folium.Icon(icon='star',color='red'), tooltip=tmp[1]).add_to(map_data)
map_data
```
